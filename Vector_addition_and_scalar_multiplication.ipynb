```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1,2,1])
k = np.array([-1,6,-1])
origin = np.array([0,0,0])

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.quiver(*origin, *v, color='r', arrow_length_ratio = 0.1)
ax.quiver(*v, *k, color='b', arrow_length_ratio = 0.1)

result = k + v
ax.quiver(*origin, *result, color='k', arrow_length_ratio = 0.1)

ax.text(
    result[0],
    result[1],
    result[2],
    "v+k",
    color="k"
)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")


ax.set_xlim(0,4)
ax.set_ylim(0,4)
ax.set_zlim(0,4)
```

```python
x = np.array([1,2,3])
y = np.array([2,-1,3])
z = np.array([3,4,1])
origin = np.array([0,0,0])

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.quiver(*origin, *x, color='r', arrow_length_ratio = 0.1)
ax.quiver(*x,*y, color='b', arrow_length_ratio = 0.1)
ax.quiver(*(x + y), *z, color='g', arrow_length_ratio=0.1)

result = x + y + z

ax.quiver(*origin, *result, color="k", arrow_length_ratio=0.1)

ax.text(
    result[0],
    result[1],
    result[2],
    "x+y+z",
    color="k"
)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")


ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.set_zlim(0,10)
```

```python
o = np.array([1,2,3])

origin = np.array([0,0,0])

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.quiver(*origin, *o, color = 'r', arrow_length_ratio=0.1)

vectox2 = 2 * o

ax.quiver(*origin, *vectox2, color='b', arrow_length_ratio=0.1)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.set_xlim(0,6)
ax.set_ylim(0,6)
ax.set_zlim(0,6)
```
